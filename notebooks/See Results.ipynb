```python
folder = 'cifar2hsi_128'
# folder = '4hsi_vanilla_version2'
epoch = '010'
filename = '../checkpoints/{0}/web/images/epoch{1}_real_B.hdf5'.format(folder, str(epoch))
with h5py.File(filename, 'r') as f:
    data = f['data']
    d1 = np.array(data)

filename = '../checkpoints/{0}/web/images/epoch{1}_fake_B.hdf5'.format(folder, str(epoch))
with h5py.File(filename, 'r') as f:
    data = f['data']
    d2 = np.array(data)
    
filename = '../checkpoints/{0}/web/images/epoch{1}_real_A.hdf5'.format(folder, str(epoch))
with h5py.File(filename, 'r') as f:
    data = f['data']
    d3 = np.array(data)

filename = '../checkpoints/{0}/web/images/epoch{1}_fake_A.hdf5'.format(folder, str(epoch))
with h5py.File(filename, 'r') as f:
    data = f['data']
    d4 = np.array(data)
    
filename = '../checkpoints/{0}/web/images/epoch{1}_rec_A.hdf5'.format(folder, str(epoch))
with h5py.File(filename, 'r') as f:
    data = f['data']
    d5 = np.array(data)

filename = '../checkpoints/{0}/web/images/epoch{1}_rec_B.hdf5'.format(folder, str(epoch))
with h5py.File(filename, 'r') as f:
    data = f['data']
    d6 = np.array(data)
```

```python
view1 = imshow(d1)
print('Visualizing the GT hyperspectral image')
print(view1)
view3 = imshow(d3)
print('Visualizing the GT hazy image')
print(view3)
view2 = imshow(d2)
print('Visualizing the generated hyperspectral image. Should be more similar to GT hazy image')
print(view2)
```

```python
view4 = imshow(d4)
print('Visualizing the generated RGB image. Should be similar to GT hyperspectral image')
print(view4)
view5 = imshow(d5)
print('Visualizing the Reconstructed RGB image.')
print(view5)
view6 = imshow(d6)
print('Visualizing the Reconstructed HSI image')
print(view6)
```
